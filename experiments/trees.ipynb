```python
def generate_swaps(perm:list[int]) -> list[tuple[int,int]]:
  arr = list(perm)
  n = len(arr)
  swaps = []

  for i in range(n):
    for j in range(0, n-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
        swaps.append((j,j+1))
  return swaps

print(generate_swaps([3, 1, 4, 2, 5, 6]))

def swap(outers, seq):
  for i,j in seq:
    outers[i],outers[j] = outers[j],outers[i]
      
```

```python
figure_50_tree = FusionTree(
  open_edges=["f"],
  internal_edges=["a", "b", "c", "d", "e"],
  nodes=[("f", "a", "e"), ("e", "b", "c"), ("a", "b", "d"), ("d", "c", "f")],
  directions=[NodeType.splitting, NodeType.splitting, NodeType.fusion, NodeType.fusion],
)

print(figure_50_tree.nodes)
figure_50_tree.fmove("d")
print(figure_50_tree.nodes)
print(figure_50_tree.directions)


def is_identity_by_labels(tree) -> bool:
  splittings = []
  fusions = []

  for node, dtype in zip(tree.nodes, tree.directions):
    if dtype == NodeType.splitting:
      splittings.append(node)
    elif dtype == NodeType.fusion:
      fusions.append(node)

  matched_pairs = 0
  for s_parent, s_c1, s_c2 in splittings:
    for f_c1, f_c2, f_parent in fusions:
      if s_parent == f_parent and ((s_c1 == f_c1 and s_c2 == f_c2) or (s_c1 == f_c2 and s_c2 == f_c1)):
        matched_pairs += 1

  return matched_pairs == len(splittings)

print(is_identity_by_labels(figure_50_tree))
```
