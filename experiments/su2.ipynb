```python
import numpy as np


def build_possible_sector(j1: float, j2: float):
    return np.arange(abs(j1 - j2), j1 + j2 + 1, 1.0)


def build_possible_sector_from_sector(sector1: list, sector2: list):
    possible_values = set()
    for j1 in sector1:
        for j2 in sector2:
            res = build_possible_sector(j1, j2)
            possible_values.update(res)

    return np.sort(np.array(list(possible_values)))


print(build_possible_sector_from_sector([0,1,2], [0,1]))
```

```python
import numpy as np
import itertools

def build_possible_sector(j1, j2):
    """Standard SU(2) triangle inequality."""
    return np.arange(abs(j1 - j2), j1 + j2 + 1, 1.0)

def get_sequential_fusion_sectors(allowed_inputs, allowed_final):
    list_of_charge_sectors = []
    
    external_inputs = list(itertools.product(allowed_inputs, repeat=3))
    
    for j1, j2, j3 in external_inputs:
        possible_j_ints = build_possible_sector(j1, j2)
        
        for j_int in possible_j_ints:
            possible_j4s = build_possible_sector(j_int, j3)
            
            for j4 in possible_j4s:
                if j4 in allowed_final:
                    list_of_charge_sectors.append([
                        int(j_int), int(j1), int(j2), int(j3), int(j4)
                    ])
                    
    return list_of_charge_sectors

inputs = [0, 1]
final_allowed = [0, 1]

sectors = get_sequential_fusion_sectors(inputs, final_allowed)
print(len(sectors))
print(sectors)
```
